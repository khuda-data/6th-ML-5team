```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

columns_to_scale = ['판매량', '매출건수']

scaled_data = scaler.fit_transform(data[columns_to_scale])

scaled_df = pd.DataFrame(scaled_data, columns=['판매량 지수', '매출건수 지수'])

result_df_standard = pd.concat([data, scaled_df], axis=1)
```

```python
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

silhouette_scores = []
inertia_values = []
cluster_range = range(2, 11)

for n_clusters in cluster_range:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(data)
    silhouette_avg = silhouette_score(data, cluster_labels)
    silhouette_scores.append(silhouette_avg)
    inertia_values.append(kmeans.inertia_)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(cluster_range, silhouette_scores, marker='o')
plt.title('Silhouette Scores for Different Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')

plt.subplot(1, 2, 2)
plt.plot(cluster_range, inertia_values, marker='o')
plt.title('Inertia (Sum of Squared Distances) for Different Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')

plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
import matplotlib.pyplot as plt

# KMeans 모델 초기화
kmeans = KMeans(random_state=42)

# KElbowVisualizer 초기화 (2에서 11까지의 클러스터 개수 탐색)
elbow_visualizer = KElbowVisualizer(kmeans, k=(2, 11))

# 데이터에 적합시키고 엘보우 그래프 그리기
elbow_visualizer.fit(data)
elbow_visualizer.show()
```

```python
optimal_kmeans = KMeans(n_clusters=4, random_state=42)

clusters = optimal_kmeans.fit_predict(data)

data['clusters'] = clusters
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
numeric_columns = data.select_dtypes(include='number')

plt.figure(figsize=(14, 10))

for i, column in enumerate(numeric_columns.columns[:-1], 1):  
    plt.subplot(3, 3, i)
    sns.boxplot(x='clusters', y=column, data=data)
    plt.title(f'Boxplot of {column} by Cluster')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.stats import f_oneway

features = numeric_columns.columns[:-1]  

anova_results = {}

for feature in features:
    clusters = [data[data['clusters'] == i][feature] for i in data['clusters'].unique()]
    
    f_stat, p_val = f_oneway(*clusters)
    
    anova_results[feature] = {'F-statistic': f_stat, 'p-value': p_val}

anova_results_df = pd.DataFrame(anova_results).T

anova_results_df = anova_results_df.sort_values(by='F-statistic', ascending=False)
```

```python
# F-통계량 값이 높을수록 클러스터를 구별하는데 중요한 역할을 한다.
anova_results_df.head(3)
```
