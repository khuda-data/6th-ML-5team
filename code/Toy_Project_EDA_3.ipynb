```python
# '의약/의료품'과 'admi_cty_no'가 41461250인 행을 필터링
consume_pharmacy = df[(df['card_tpbuz_nm_2'] == '의약/의료품')]

mapping = {
    41463516: '영덕동', 41463517: '영덕동',
    41463572: '동백동', 41463575: '동백동', 41463577: '동백동',
    41465510: '풍덕천동', 41465520: '풍덕천동',
    41465540: '죽전동', 41465550: '죽전동', 41465555: '죽전동',
    41465570: '상현동', 41465580: '상현동', 41465585: '상현동',
    41461250: '포곡읍', 41461253: '모현읍', 41461256: '이동읍', 41461259: '남사읍',
    41461340: '원삼면', 41461350: '백암면', 41461360: '양지면',
    41461510: '중앙동', 41461525: '역삼동', 41461526: '역삼동',
    41461530: '유림동', 41461540: '동부동', 41463510: '신갈동',
    41463520: '구갈동', 41463530: '상갈동', 41463535: '보라동',
    41463540: '기흥동', 41463550: '서농동', 41463560: '구성동',
    41463570: '마북동', 41463586: '상하동', 41463590: '보정동',
    41465530: '신봉동', 41465560: '동천동', 41465590: '성복동'
}

# 매핑 적용
consume_pharmacy['읍면동'] = consume_pharmacy['admi_cty_no'].map(mapping)

# 읍면동별 매출건수 합산
sales_sum = consume_pharmacy.groupby('읍면동', as_index=False)['cnt'].sum().rename(columns={'cnt': '매출건수'})

# 읍면동별 판매량(행의 개수) 계산
sales_count = consume_pharmacy.groupby('읍면동').size().reset_index(name='판매량')

# 두 DataFrame 병합
final_result = pd.merge(sales_count, sales_sum, on='읍면동')

# csv 파일로 저장
final_result.to_csv('final_result.csv', index=False)

final_result
```
