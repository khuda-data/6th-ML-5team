```python
import folium
import numpy as np
import geopandas as gpd
import matplotlib.colors as mcolors
from shapely.geometry import box
import pandas as pd

# 용인시 중심 좌표 계산
center = yongin.geometry.centroid.iloc[0]

# folium 지도 생성 (중심점을 용인시 중심으로 설정)
m = folium.Map(location=[center.y, center.x], zoom_start=11, tiles='OpenStreetMap')

# 용인시 경계 추가
folium.GeoJson(yongin, name="Yongin Boundary").add_to(m)

# Choropleth 지도 생성
choropleth = folium.Choropleth(
    geo_data=yongin,
    name="choropleth",
    data=final_data,
    columns=["읍면동", "접근성 지수"],  # Clusters 열을 사용하여 색상 지정
    key_on="feature.properties.EMD_KOR_NM",  # GeoJSON 파일의 키와 매칭
    fill_color="Blues",
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="접근성 지수"
).add_to(m)

# GeoJSON 레이어 추가
folium.GeoJson(
    yongin,
    style_function=lambda feature: {
        'fillColor': 'transparent',
        'color': 'black',
        'weight': 1
    }
).add_to(m)

# 용인시 경계를 따라 격자 추가
lat_min, lat_max = yongin.total_bounds[1], yongin.total_bounds[3]
lon_min, lon_max = yongin.total_bounds[0], yongin.total_bounds[2]
lat_steps = 40  # 위도 방향으로 격자선 수를 증가시켜 더 촘촘하게
lon_steps = 40  # 경도 방향으로 격자선 수를 증가시켜 더 촘촘하게

# 용인시 경계
yongin_boundary = yongin.unary_union  # 용인시 전체 경계를 하나의 객체로 결합

# 격자를 용인시 경계에 맞춰서 그리기
for lat in np.linspace(lat_min, lat_max, lat_steps):
    for lon in np.linspace(lon_min, lon_max, lon_steps):
        grid_cell = box(lon, lat, lon + (lon_max-lon_min)/lon_steps, lat + (lat_max-lat_min)/lat_steps)
        intersection = grid_cell.intersection(yongin_boundary)
        if not intersection.is_empty:
            folium.GeoJson(
                gpd.GeoSeries([intersection]).__geo_interface__,
                style_function=lambda feature: {
                    'fillColor': 'transparent',
                    'color': 'gray',
                    'weight': 1.5,
                    'fillOpacity': 0
                }
            ).add_to(m)

# 엑셀 데이터 로드 및 점 추가
data = pd.read_excel('/content/drive/MyDrive/KHUDA/ML/ML toy project/용인시 폐의약품 수거함 위도경도.xlsx')

# 지도에 빨간 점 추가
for idx, row in data.iterrows():
    folium.CircleMarker(
        location=(row['latitude'], row['longtitude']),
        radius=5,
        color='red',
        fill=True,
        fill_color='red',
        fill_opacity=0.8
    ).add_to(m)

# 지도를 HTML 파일로 저장하고 표시
m.save("yongin_map.html")

# 지도 객체 반환
m
```

```python
import folium
import numpy as np
import geopandas as gpd
import pandas as pd

# Yongin city center coordinates (centroid)
center = yongin.geometry.centroid.iloc[0]

# Create a folium map centered on Yongin city
m = folium.Map(location=[center.y, center.x], zoom_start=11, tiles='OpenStreetMap')

# Add Yongin city boundary
folium.GeoJson(yongin, name="Yongin Boundary").add_to(m)

# Choropleth map - Accessibility Index
choropleth1 = folium.Choropleth(
    geo_data=yongin,
    name="접근성 지수",
    data=final_data,
    columns=["읍면동", "접근성 지수"],
    key_on="feature.properties.EMD_KOR_NM",
    fill_color="Blues",
    fill_opacity=0.3,  # Increase transparency
    line_opacity=0.2,
    legend_name="접근성 지수"
).add_to(m)

# Choropleth map - Number of Routes
choropleth2 = folium.Choropleth(
    geo_data=yongin,
    name="노선수",
    data=final_data,
    columns=["읍면동", "노선수"],
    key_on="feature.properties.EMD_KOR_NM",
    fill_color="Greens",
    fill_opacity=0.3,  # Increase transparency
    line_opacity=0.2,
    legend_name="노선수"
).add_to(m)

# Choropleth map - Sales
choropleth3 = folium.Choropleth(
    geo_data=yongin,
    name="판매량",
    data=final_data,
    columns=["읍면동", "판매량"],
    key_on="feature.properties.EMD_KOR_NM",
    fill_color="Purples",
    fill_opacity=0.3,  # Increase transparency
    line_opacity=0.2,
    legend_name="판매량"
).add_to(m)

# Add GeoJSON layer
folium.GeoJson(
    yongin,
    style_function=lambda feature: {
        'fillColor': 'transparent',
        'color': 'black',
        'weight': 1
    }
).add_to(m)

# Adding grid lines along Yongin city boundary
lat_min, lat_max = yongin.total_bounds[1], yongin.total_bounds[3]
lon_min, lon_max = yongin.total_bounds[0], yongin.total_bounds[2]
lat_steps = 50  # Increase the number of grid lines in the latitude direction
lon_steps = 50  # Increase the number of grid lines in the longitude direction

# Yongin city boundary
yongin_boundary = yongin.unary_union  # Combine Yongin city boundary into a single object

# Draw the grid according to Yongin city boundary
for lat in np.linspace(lat_min, lat_max, lat_steps):
    for lon in np.linspace(lon_min, lon_max, lon_steps):
        grid_cell = box(lon, lat, lon + (lon_max-lon_min)/lon_steps, lat + (lat_max-lat_min)/lat_steps)
        intersection = grid_cell.intersection(yongin_boundary)
        if not intersection.is_empty:
            folium.GeoJson(
                gpd.GeoSeries([intersection]).__geo_interface__,
                style_function=lambda feature: {
                    'fillColor': 'transparent',
                    'color': 'gray',
                    'weight': 1.5,
                    'fillOpacity': 0
                }
            ).add_to(m)

# Load Excel data
data = pd.read_excel('/content/drive/MyDrive/KHUDA/ML/ML toy project/버스스탑.xlsx')

# Add blue markers for the coordinates on the map
for idx, row in data.iterrows():
    folium.Marker(
        location=(row['latitude'], row['longtitude']),
        icon=folium.Icon(color='blue')
    ).add_to(m)

# Add layer control
folium.LayerControl().add_to(m)

# Save the map as an HTML file and display
m.save("yongin_map_with_blue_markers.html")

# Return the map object
m
```
