```python
data = pd.merge(people_data, bus_data, on=['시군구', '읍면동'], how='left')
```

```python
data = data[['시도', '시군구', '읍면동', '등록인구', '65세이상 고령자', '65세미만 인구', '노선수']]
```

```python
data
```

```python
from sklearn.preprocessing import StandardScaler

numerical_columns = data.select_dtypes(include=['int64', 'float64']).columns

scaler = StandardScaler()

numerical_columns = ['등록인구', '65세이상 고령자', '65세미만 인구', '노선수']
scaled_data = scaler.fit_transform(data[numerical_columns])

scaled_df = pd.DataFrame(scaled_data, columns=numerical_columns)
```

```python
data['접근성 지수'] = (data['등록인구'] * 0.3) + (data['65세이상 고령자'] * 0.3) + (data['노선수'] * 0.4)
```

```python
data = pd.concat([bus_data[['시도', '시군구', '읍면동']], scaled_df], axis=1)
```
